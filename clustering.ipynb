```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set_style('darkgrid')
%matplotlib inline
```

```python
flowerdf = sns.load_dataset('iris')
```

```python
flowerdf
```

```python
sns.scatterplot(data=flowerdf, x='sepal_length', y='petal_length', hue='species');
```

```python
numeric_cols = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
```

```python
y = flowerdf[numeric_cols]
```

```python
model = KMeans(n_clusters=3, random_state=69)
```

```python
model.fit(y)
```

```python
preds = model.predict(y)
preds
```

```python
sns.scatterplot(data=y, x='sepal_length', y='petal_length', hue=preds);
centers_x, centers_y = model.cluster_centers_[:,0], model.cluster_centers_[:,2]
plt.plot(centers_x, centers_y, 'xb')
```

```python
model.inertia_
```

```python
model2 = KMeans(n_clusters=6, random_state=420)
```

```python
model2.fit(y)
```

```python
preds2 = model2.predict(y)
preds2
```
